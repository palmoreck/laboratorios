# Ejemplo de transformaciones lineales: proyección, reflexión y rotación en el plano

Un conjunto de vectores $\{x_1, \dots, x_p\}$ en $\mathbb{R}^m$ ($x_i \in \mathbb{R}^m$)es ortogonal si $x_i^Tx_j=0$ $\forall i\neq j$. Por ejemplo, para un conjunto de $2$ vectores $x_1,x_2$ en $\mathbb{R}^3$ esto se visualiza:

<img src="https://dl.dropboxusercontent.com/s/cekagqnxe0grvu4/vectores_ortogonales.png?dl=0" heigth="550" width="550">


## Matriz ortogonal

Una matriz $A \in \mathbb{R}^{nxn}$ se nombra ortogonal si sus columnas son ortogonales y cada columna es ortonormal. Tal propiedad se puede escribir de forma matricial como $A^TA = I$ donde: $I$ es la matriz identidad.

Como ejemplos de matrices ortogonales en $\mathbb{R}^{nxn}$ se tienen las matrices de rotación y reflexión.

### Matriz de rotación en sentido contrario a las manecillas del reloj en el plano

Está dada por: $$R_{\theta} = \begin{bmatrix} cos(\theta) & -sen(\theta) \\ sen(\theta)& cos(\theta)\end{bmatrix}$$

Usar esta matriz de rotación para rotar un ángulo de $\frac{\pi}{3}$ en sentido contrario a las manecillas del reloj al vector $v = \left[\begin{array}{c}
-3\\
4\\
\end{array}
\right]$.

```python
import math
```

Definición de $v$:

```python
v = np.array([-3,4.0])
```

Definición de matriz $R_\theta$ con $\theta = \pi/3$:

```python
theta = math.pi/3
R_theta = np.array([[math.cos(theta), -math.sin(theta)],
                    [math.sin(theta), math.cos(theta)]])
```

```python
R_theta
```

Realizamos la rotación para obtener $v_{\text{rot}}$ para esto realizamos la multiplicación $R_\theta v$:

```python
v_rot = R_theta@v
```

```python
v_rot
```

Realizamos la gráfica:

```python
t1 = (np.zeros(2), v)

line_seg_v = np.row_stack(t1)
```

```python
line_seg_v
```

```python
t2 = (np.zeros(2), v_rot)

line_seg_v_rot = np.row_stack(t2)
```

```python
line_seg_v_rot
```

```python
plt.plot(line_seg_v[:,0], line_seg_v[:,1], 'b-',
         line_seg_v_rot[:,0], line_seg_v_rot[:,1],'r-')
plt.scatter(v[0], v[1], color="blue")
plt.scatter(v_rot[0], v_rot[1], color="red")
plt.grid()
plt.axhline(color="black")
plt.axvline(color="black")
plt.legend(['Vector', 'Vector con rotación'])
plt.title("Hacemos rotación de $\pi/3$")
plt.show()
```

## Matriz de proyección elemental ortogonal

Está dada por: $$P = I - \frac{uu^T}{u^Tu}$$ con $u \in \mathbb{R}^n -\{0\}$ y realiza la proyección en el conjunto de vectores ortogonales a $u$. 

Usar esta matriz de proyección elemental ortogonal para proyectar al vector $v = \left[\begin{array}{c}
-0.4\\
-0.8\\
\end{array}
\right]$ utilizando al vector $ u = \left[\begin{array}{c}
-1\\
\frac{1}{3}\\
\end{array}
\right]$ para construir a $P$

```python
v = np.array([-.4, .8])
```

```python
u = np.array([-1, 1/3])
```

```python
P = np.eye(2) - np.outer(u, u)/u.dot(u)
```

```python
P
```

```python
v_proj = P@v
```

```python
v_proj
```

```python
t1 = (np.zeros(2), v)
t2 = (np.zeros(2), v_proj)
t3 = (np.zeros(2), u)
line_seg_v = np.row_stack(t1)
line_seg_v_proj = np.row_stack(t2)
line_seg_u = np.row_stack(t3)
```

```python
line_seg_v
```

```python
line_seg_v_proj
```

```python
line_seg_u
```

```python
x_plot = np.linspace(-2, 2, 100)
```

```python
plt.figure(figsize=(5, 7))
plt.plot(line_seg_v[:,0], line_seg_v[:,1], 'b-',
         line_seg_v_proj[:,0], line_seg_v_proj[:,1], 'r-',
         line_seg_u[:,0], line_seg_u[:,1], 'g-',
         x_plot*v_proj[0], x_plot*v_proj[1], 'r--')
plt.scatter(v[0], v[1], color = 'b')
plt.scatter(v_proj[0], v_proj[1], color = 'r')
plt.scatter(u[0], u[1], color = 'g')
plt.legend(["vector", "vector proyectado",
            "vector para proyección u", "complemento ortogonal de u"], bbox_to_anchor=(1, 1))
plt.axhline(color="black")
plt.axvline(color="black")
plt.title("Ejemplo proyector ortogonal elemental")
plt.grid()
plt.show()
```
