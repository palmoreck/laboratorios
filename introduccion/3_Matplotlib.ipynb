# Figure, axes

Una figura nos proporciona un lienzo, *canvas*, para la graficación. Los ejes, *axes*, proveen un sistema de coordenadas en el *canvas*. La figura y los *axes* son objetos que tienen métodos para configurar y personalizar propiedades de la gráfica.

```python
x = np.linspace(-5, 2, 100)
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10*x
y3 = 6*x + 10
```

```python
fig, axes = plt.subplots()
axes.plot(x, y1, color="blue", label="y(x)")
axes.plot(x, y2, color="red", label="y'(x)")
axes.plot(x, y3, color="green", label="y''(x)")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.legend()
plt.show()
```

# Ejemplo: rotación en el plano

Definición de $v$:

```python
v = np.array([-3,4.0])
```

Definición de matriz $R_\theta$ con $\theta = \pi/3$:

```python
theta = math.pi/3
R_theta = np.array([[math.cos(theta), -math.sin(theta)],
                    [math.sin(theta), math.cos(theta)]])
```

Realizamos la rotación para obtener $v_{\text{rot}}$:

```python
v_rot = R_theta@v
```

```python
v_rot
```

Realizamos la gráfica:

```python
A = np.array([[0, 0],
              v])
```

```python
A
```

```python
A2 = np.array([[0, 0],
               v_rot])
```

```python
A2
```

```python
plt.plot(A[:,0], A[:,1], '-',
         A2[:,0], A2[:,1],'r-',
         )
plt.legend(['Vector', 'Vector con rotación'])
plt.title("Hacemos rotación de $\pi/3$")
plt.show()
```
